```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from bs4.dammit import EncodingDetector
from selenium.webdriver.common.action_chains import ActionChains
import os
import random as rd
import time

def conectarConSitio(url="https://www.academico.espol.edu.ec/"):
    driver = webdriver.Chrome(executable_path=r"chromedriver.exe")
    driver.get(url)
    return driver

def cambiarDePestana(url):
    driver.get(url)

def llenarConsultaAnioyBuscar():
    idinputAnio = "ctl00_contenido_txtAnio"
    idinputTermino = "ctl00_contenido_listab_1"
    idbotonConsultar = "ctl00_contenido_btnConsultar"
    
    driver.find_element_by_id(idinputAnio).send_keys("2020")
    inputTermino = driver.find_element_by_id(idinputTermino)
    botonConsultar = driver.find_element_by_id(idbotonConsultar)

    actions = ActionChains(driver)
    actions.click(inputTermino)
    actions.click(botonConsultar)
    actions.perform()
def obtenerElementosBusquedaMateria():
    #retorna el input de nombre materia y el boton de buscar(para hacer click())
    idInputNombreMateria = "ctl00_contenido_nombreMateria"
    nombreInputNombreMateria = "ctl00$contenido$nombreMateria"
    idBotonBuscar = "ctl00_contenido_Button2"
    
    inputNombreMateria = driver.find_element_by_name(nombreInputNombreMateria)
    botonBuscar = driver.find_element_by_id(idBotonBuscar)
    return inputNombreMateria,botonBuscar

def escribirArchivo(document, nombre):
    archivo = open(nombre,"a",encoding = "utf-8")
    archivo.write(document)
    archivo.close()
def crearDirectorio(directoryName,codigo):
    try:
        os.mkdir("Materias/"+directoryName+"_"+codigo)
    except:
        print("Ya existe un directorio con : {} ,como nombre".format(directoryName))
def obtenerNombreCompleto(directoryName, aName):
    return "Materias/"+directoryName+"/"+directoryName+"-"+aName+".html"
```

```python
def buscarMateria(nombreMateria, entradaTexto, boton ):
    entradaTexto.clear()
    entradaTexto.send_keys(nombreMateria)
    actions = ActionChains(driver)
    actions.click(boton)
    actions.perform()
```

```python
#"registros.txt"
def obtenerDicCodigos(archivo):
    archivoMaterias = open(archivo,"r",encoding = "utf-8")
    dicMaterias = {}
    lineas = 0
    for mat in archivoMaterias:
        lineas+=1
        codigo, nombre, paralelos = mat.strip().split("|")
        paralelos = paralelos.replace("[","").replace("]","").split(",")
        paralelos = [str(int(par.replace("'",""))) for par in paralelos]
        sub = dicMaterias.get(codigo, dict())
        listaParalelos = sub.get(nombre, list())
        listaParalelos.extend(paralelos)
        sub[nombre] = listaParalelos
        dicMaterias[codigo] = sub
    archivoMaterias.close()
    return dicMaterias
```

```python
def revertDic(dicOriginal):
    dicReverse = {}
    for codigo, sub in dicOriginal.items():
        for name in sub:
            listaCodigos = dicReverse.get(name,[])
            listaCodigos.append(codigo)
            dicReverse[name] = listaCodigos
    return dicReverse
```

```python
def mostrarDic(dicReverse):
    for name,codigo in dicReverse.items():
        print("{} #{} , {}".format(name,len(codigo),",".join(codigo)))
```

```python
dicMat = obtenerDicCodigos("registros2.txt")
dicRev = revertDic(dicMat)
print("Deberia existir {} directorios".format(len(dicRev)))
#mostrarDic(dicRev)
```

```python
encontrados = 0
for nombre, listCodigo in dicRev.items():
    if os.path.exists(nombre):
        codigo = listCodigo[0]
        nuevoNombre = nombre + "_" + codigo
        os.rename(nombre,nuevoNombre)
    else:
        encontrados += 1
        print("No se encontro el directorio con el nombre: {}".format(nombre))
print("No se encontraron {} dir".format(encontrados))
```
